FORTNITE TOP 500 PLAYERS DATA ANALYSIS

The purpose of this analysis is to discover the the countires with the greatest Fortnite Culture on the basis of the top 500 players.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('fortnite500.csv', encoding='Latin-1')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
by_comp = df.groupby(by=' Country')
```

```python
len(list(df[' Country'].unique()))-1    #Number of Countires, -1 bec Nan is counted
```

```python
df['wg'] = df[' Wins']/df[' Games']
```

```python
df.head()
```

```python
df['wl']=df[' Wins']/(df[' Games']-df[' Wins'])
```

```python
df.head()
```

```python
df.sort_values('wl',ascending=False).head(10)
```

```python
df.sort_values('wg',ascending=False).head()
```

```python
dfdropna = df.dropna()
dfdropna.head()
```

```python
dfdropna.head()
dfdropna.columns = ['Rank','Name','Platform','Country','Wins','Games','wg','wl']
dfdropna.head()
```

```python
nation = dfdropna.groupby('Country')
```

```python
nation.describe
```

```python
by_count = dfdropna.groupby('Country')
```

```python
num = list(by_count.agg('count')['Wins'])
```

```python
del num[-1]
num
```

```python
count = list(by_count.groups.keys())
del count[-1]
count
```

```python
import pycountry
```

```python
from pprintpp import pprint

from pycountry_convert import (
    map_countries,
    country_alpha2_to_country_name,
    country_name_to_country_alpha2,
    country_alpha3_to_country_alpha2,
    country_name_to_country_alpha3,

    country_alpha2_to_continent_code,

    COUNTRY_NAME_FORMAT_UPPER
)
cn_name_format = COUNTRY_NAME_FORMAT_UPPER
```

```python
country = []
for code in count:
    cn_name = country_alpha2_to_country_name(code, cn_name_format)
    countryy = country_name_to_country_alpha3(cn_name, cn_name_format)
    country.append(countryy)
```

```python
country
```

```python
map_data = pd.DataFrame({'Country': country,'z': num})
```

```python
map_data
```

```python
import plotly.graph_objs as go 
from plotly.offline import init_notebook_mode,iplot
init_notebook_mode(connected=True) 
```

```python
data = dict(type = 'choropleth',
           locations = map_data['Country'],
           z = map_data['z'],
           colorbar = {'title' : 'Number of Players'},
           text = map_data['Country'],
           colorscale = 'Viridis',
           locationmode = "country abbreviation",
           reversescale = 'True')

layout = dict(title = 'Fortnite Top Players',
             geo = dict(showframe = False,projection = {'type':'Mercator'}))
```

```python
choromap = go.Figure(data = [data],layout = layout)
iplot(choromap,validate=False)
```

```python
total_dubs = list(by_count.sum()['Wins'])
```

```python
total_dubs
del total_dubs[-1]
```

```python
tdubs = pd.DataFrame({'Country': country,'z': total_dubs}) 
```

```python
tdubs
```

```python
import seaborn as sns

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(25,10))
sns.barplot(x='Country',y='z',data = tdubs)
```

Through the choropleth map and the barplot, the countries with the strongest Fortnite Culture can be easily found. This analysis was an attempt to convert the subjective answer to the question of the strenght of Fortnite Culture in to the objective expression of comparable numericals on the basis of top 500 players.
